```python
import pandas as pd
from sqlalchemy import create_engine, text
from common import config
# 连接数据库
engine = create_engine(url=config.DB_URL, connect_args={'options': '-c timezone=Asia/Shanghai'})
# 创建游标
df = pd.read_sql("SELECT * FROM fund_market where symbol='159117.SZ'", engine)
print(df)
# 关闭连接
engine.dispose()
```
